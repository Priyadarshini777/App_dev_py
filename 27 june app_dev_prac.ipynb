```python
"""Use Case: Admin User Managing Documents
Description:
Admin users can perform all CRUD (Create, Read, Update, Delete) operations on any documents in the system."""
# A list to store documents

documents = []

#  admin credentials stored in a dictionary
admin_credentials = {
    'username': 'Priya',
    'password': 'Priya123'
}

#function to check whether the input given by the user is valid or not accordingly to the records stored
def login(username, password):
    if username == admin_credentials['username'] and password == admin_credentials['password']:
        print("Admin logged into the system.")
        return True
    else:
        print("Invalid credentials. Login failed.")
        return False
#function to represent whether admin has logged in and ready to move further
def navigate_to_document_management():
    print("Admin navigated to the document management section.")

#create function
def create_document(content):
    documents.append(content)
    print(f"Document created with content: {content}")

#update function
def update_document(index, new_content):
    if 0 <= index < len(documents):
        documents[index] = new_content
        print(f"Document at index {index} updated with new content: {new_content}")
    else:
        print(f"Document at index {index} not found. Update failed.")

#delete function
def delete_document(index):
    if 0 <= index < len(documents):
        documents.pop(index)
        print(f"Document at index {index} deleted.")
    else:
        print(f"Document at index {index} not found. Deletion failed.")

#retrieve function
def view_document(index):
    if 0 <= index < len(documents):
        print(f"Document at index {index}: {documents[index]}")
    else:
        print(f"Document at index {index} not found.")



# Admin login
username = input("Enter username: ")
password = input("Enter password: ")

if login(username, password):
    # Admin navigates to the document management section
    navigate_to_document_management()

    # Create documents
    create_document("First document content")
    create_document("Second document content")

    # Update a document
    update_document(0, "Updated first document content")

    # View documents
    view_document(0)
    view_document(1)

    # Delete a document
    delete_document(1)

    # Try to view a deleted document
    view_document(1)

```
