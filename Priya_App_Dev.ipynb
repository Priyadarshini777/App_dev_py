```python
# Constants
# INITILISiNG THE DATA BASE RECORD
data = {"records": []}
EXCLUDED_NAMES = ["John Doe", "Jane Doe", "Admin"]
#the above list contains all the names tht are conditionally not included for validation

```

```python
#util :validation functions for name length, type, and characters.
def is_excluded(name):#function to define the rules for the excluded names
    if name in EXCLUDED_NAMES:
        print(f"{name} is in the excluded list")
        print("Name verification is successful")
        return True#boolean result
    return False

def is_valid_length(name): #to check the length and set constraints for the length of the name input
    if 2 <= len(name) <= 50:
        return True
    else:
        raise ValueError(f"Invalid name length {len(name)}; valid length is between 2 and 50 characters") #if not will raise the value error

def is_valid_type(name):
    if isinstance(name, str):# function to check if the input is string or not
        return True
    else:
        raise ValueError(f"Invalid name type - {type(name)}") #once error returns the correct datatype of the input

def is_valid_characters(name):
    if all(char.isalpha() or char.isspace() for char in name): #isalpha and isspace function to check and set constraint as the input should only contain alpahabets and space
        return True
    else:
        raise ValueError(f"Invalid characters in name - {name}")
```

```python
#function checks if a name is valid based on the new validation rules
def is_valid_name(name): #function including alll the above functions and checking the validation of the name
    if is_valid_type(name) and is_valid_length(name) and is_valid_characters(name):
        # Excluded persons no further validation required
        if is_excluded(name):
            return True
        print("Name verification is successful")
        return True
    return False

# Example usage
#exception handling concept implementation and calling the function with input parameters
try:
    print(is_valid_name("John Doe"))  # This should be excluded
    print(is_valid_name("Alice"))     # This should be valid
    print(is_valid_name("123"))       # This should raise a ValueError for invalid characters
except ValueError as e:
    print(e)
```
